Q1.[10 points] 
---
In this question, you have to create a database, create a table, insert two rows of data and fetch all rows and return that list data. Information on the same is as below:  
database name : 'lab11.db'  
table name : 'sample_table'  
    - columns in the table : (name(text), age(integer), address(text))  

first row data : ('sherlock holmes', 33, '221 B Baker street')  
second row data : ('harry potter', 45, 'station 9 3/4, london') 

Note: You may want to drop the table before you create it.

---
Grading Rubric: Binary

```python
import sqlite3


def sample_data():
    connection = sqlite3.connect("lab11.db")
    cur = connection.cursor()
    
    cur.execute("DROP TABLE IF EXISTS sample_table")
    
    cur.execute("CREATE TABLE sample_table (name text, age integer, address text)")
    
    cur.execute("INSERT INTO sample_table (name,age,address) VALUES ('sherlock holmes', 33, '221 B Baker street'), ('harry potter', 45, 'station 9 3/4, london')")
    
    connection.commit()
    
    cur.execute("SELECT * FROM sample_table")
    vals = cur.fetchall()

    cur.close()
    
    return vals
    pass
sample_data()
```

```python
import Lab6Code
def get_titanic_dataset():
    titanic_data = Lab6Code.csv_dataset("titanic_data.csv", ",")
    l = titanic_data.__len__()
    
    connection = sqlite3.connect("lab11.db")
    cur = connection.cursor()
    
    cur.execute("DROP TABLE IF EXISTS titanic_dataset")
    cur.execute("CREATE TABLE titanic_dataset (PassengerId Integer PRIMARY KEY , Survived Integer, Pclass Integer, Name Text, Sex Text, Age Integer, Ticket Text, Fare Real)")
    
    for i in range(1,l):
        PassengerID = titanic_data.__getitem__(i)[0]
        Survived = titanic_data.__getitem__(i)[1]
        PClass = titanic_data.__getitem__(i)[2]
        Name = titanic_data.__getitem__(i)[3]
        Sex = titanic_data.__getitem__(i)[4]
        Age = titanic_data.__getitem__(i)[5]
        Ticket = titanic_data.__getitem__(i)[6]
        Fare = titanic_data.__getitem__(i)[7]
        
        cur.execute("INSERT INTO titanic_dataset (PassengerId, Survived, Pclass, Name, Sex, Age, Ticket, Fare) VALUES (?,?,?,?,?,?,?,?)",(PassengerID, Survived, PClass, Name, Sex, Age, Ticket, Fare))
        connection.commit()
        
    result = cur.execute("SELECT * FROM titanic_dataset where PassengerId is 75")
    
    for row in result:
        res = row
        break
    cur.close()
    
    return res
    pass
get_titanic_dataset()
print(get_titanic_dataset())
```
