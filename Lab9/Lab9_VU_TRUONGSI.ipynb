````python
# Uncomment following line to install beautifulsoup if you find out that you don't have it
# !pip install bs4
import re
from bs4 import BeautifulSoup
import requests

def get_news_of(keyword):
    results = []
    content = requests.get('https://www.dailynews.com/')
    soup = BeautifulSoup(content.content, 'html.parser')
    
    tags = soup.find_all('a', {"class" : "article-title"})
    for tag in tags:
        if re.search(keyword, repr(tag.get("href")), re.IGNORECASE) or re.search(keyword, repr(tag.get_text().strip()), re.IGNORECASE):
            results.append((tag.get("href"),tag.get_text().strip()))
      
    return results
    pass
````

````python
get_news_of('Snyder')
get_news_of('Coronavirus')
````

````python
def get_airport_info():
    file = open('Airports.txt', encoding='utf8')
    values = dict()
    cnt = 0
    for line in file:
        airport_name, coordinates = line.split('\t')[1], line.split('\t')[3].split(',')
        long, lat = float(coordinates[0][1:]), float(coordinates[1][1:-2])
        values[airport_name] = (lat, long)
    return values
````

````python
get_airport_info()
````

````python
import requests
import json
def to_fly_or_not(airport_name):
    airports = get_airport_info()
    lat = str(airports.get(airport_name)[0])
    lng = str(airports.get(airport_name)[1])

    url = "https://api.weather.gov/points/" + lat + "," + lng
    r = requests.get(url)
    hourlyForecastURL = r.json()["properties"]["forecastHourly"]
    hourlyForecastRequest = requests.get(hourlyForecastURL)
    
    
    periods = hourlyForecastRequest.json()["properties"]["periods"]
    
    count = 3
    for period in periods:
        count = count - 1
        if (count <= 0):
            break
        if "cloudy" in period["shortForecast"] or "Cloudy" in period["shortForecast"]:
            return False
        
    return True

    pass

to_fly_or_not('Los Angeles County Sheriff\'s Department Heliport')
````

## Answer

Suppose every day when we crawl the headlines, we read all of them, and every article in the following day are new, we can store a list of articles that we have read by their ids, which are accessible through the property "data-id" of the "article" class. 

```html
<article class=" feature-small archive-view has-image post-4720385 post type-post status-publish format-standard has-post-thumbnail hentry category-los-angeles-dodgers category-mlb category-sports type-of-work-news" data-id="4720385" data-timestamp="1634075442">
    
``` 
Also, since each article also has a "data-timestamp" property, which can be converted to real human date and time, we can use the timestamp to crawl only the articles that were posted after some time (eg: today, tomorrow, some random date in the future)

The psuedocode for this question would be something like

* for tag in tags: if conversion_function(tag.get("data-timestamp")) older than some_date_we_want: don't crawl it
